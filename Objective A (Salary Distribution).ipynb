```python
import ast
import numpy as np
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  
import seaborn as sns
from adjustText import adjust_text

df = pd.read_csv('/Users/andreawei/Documents/Learnings/Python Project/Datasets/jobs_data_final.csv')
df['date_time'] = pd.to_datetime(df['date_time'])
df = df.dropna(subset='salary_standardized')
```

```python
sns.displot(df['salary_standardized'], kind='kde', fill=True, common_norm=False)

plt.title('Distribution of United States Data Analyst Annual Salaries (Late 2022 to Early 2025)')
plt.xlabel('Annual Salary (USD)')
plt.ylabel('Density')
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda y, pos: f'${int(y/1000)}K'))
plt.show()
```
